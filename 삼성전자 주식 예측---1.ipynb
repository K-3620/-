```python
# target = ssss_df_scaled['close']
target = ssss['close']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ssss.drop('close', axis=1, inplace=True)
```

```python
ssss
```

```python
scaler = StandardScaler()
scaler.fit(ssss)
ssss_scaled = scaler.transform(ssss)
```

```python
ssss_scaled
```

```python
ssss_df_scaled = pd.DataFrame(data=ssss_scaled,columns=ssss.columns)
```

```python
ssss_df_scaled
```

```python
train = ssss_df_scaled
```

```python
X_train, X_test, y_train, y_test =\
train_test_split(
    train
    , target
    , test_size =0.2
    , random_state = 156
)
```

```python
np.array(X_train).shape, np.array(y_train).shape
```

```python
from keras.models import Sequential
from keras.layers import Dense

layers = []
layers.append(
    Dense(
        units=128
        , input_dim=25
        , activation='relu' 
    )
)

layers.append(
    Dense(
        units= 64
        , activation = 'relu'
        , trainable=True
    )
)

layers.append(
    Dense(
        units= 32
        , activation = 'relu'
        , trainable=True
    )
)
layers.append(
    Dense(
        units= 1
        , activation = 'relu'
        , trainable=True # 학습을 저장하겠다
    )
)

model = Sequential(layers)
model.compile(loss='mse', optimizer='adam')
```

```python
history = model.fit(X_train,y_train,epochs=100,batch_size=10)
```
